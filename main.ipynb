```python
def gethastags(data):
    hashtags_list = []
    hashtags = pd.DataFrame(raw,columns=["hashtags"])
    dic_ht = hashtags.to_dict()
    dic_ht = dic_ht["hashtags"]
    for i in dic_ht:
        for j in dic_ht[i]:
            hashtags_list.append(j.lower())
    tmp = [0] * len(set(hashtags_list))
    return dict(zip( set(hashtags_list),tmp))
hastags = gethastags(raw)
```

```python
def getdataM(data,M):
    data = pd.DataFrame(data,columns=["created_at","full_text"])
    tmp = pd.Timestamp(("2020-0"+str(M)+"-01"),tz="UTC")
    tmp_1 = pd.Timestamp(("2020-0"+str(M+1)+"-01"),tz="UTC")
    data = data[data.created_at < tmp_1]
    return data[data.created_at > tmp]
```

```python
getdataM(raw,5)
```

```python
data = getdataM(raw,3)
for i in data.index:
    text = data.at[i,'full_text'].split()
    for j in text:
        if j.lower() in hastags:
            hastags[j.lower()] += 1

print(hastags)
```
